```python
# Hàm chuyển đổi ObjectId thành chuỗi
def convert_objectid(obj):
    if isinstance(obj, ObjectId):
        return str(obj)
    elif isinstance(obj, dict):
        return {k: convert_objectid(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [convert_objectid(i) for i in obj]
    else:
        return obj
```

```python
# Lấy tất cả dữ liệu từ các collection
personal_info_data = collection_personal_info.find()
insurance_info_data = collection_insurance_info.find()
query_info_data = collection_query_info.find()

# In dữ liệu thông tin cá nhân
print("Thông Tin Cá Nhân:")
personal_info_list = list(personal_info_data)
print(json.dumps(convert_objectid(personal_info_list), indent=4, ensure_ascii=False))
print("\n")  # Xuống dòng sau phần thông tin cá nhân

# In dữ liệu thông tin bảo hiểm
print("Thông Tin Bảo Hiểm:")
insurance_info_list = list(insurance_info_data)
print(json.dumps(convert_objectid(insurance_info_list), indent=4, ensure_ascii=False))
print("\n")  # Xuống dòng sau phần thông tin bảo hiểm

# In dữ liệu truy vấn cư trú
print("Truy Vấn Cư Trú:")
query_info_list = list(query_info_data)
print(json.dumps(convert_objectid(query_info_list), indent=4, ensure_ascii=False))
print("\n")  # Xuống dòng sau phần truy vấn cư trú
```

```python
# Function to convert ObjectId to string
def s_convert_objectid(obj):
    if isinstance(obj, ObjectId):
        return str(obj)
    elif isinstance(obj, dict):
        return {k: convert_objectid(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [convert_objectid(i) for i in obj]
    else:
        return obj

# Sử dụng aggregation để kết nối các collection
pipeline = [
    {
        "$lookup": {
            "from": "luuthongtinbaohiem",
            "localField": "MaThe",
            "foreignField": "MaThe",
            "as": "insurance_info"
        }
    },
    {
        "$lookup": {
            "from": "truyvetcutru",
            "localField": "MaThe",
            "foreignField": "MaThe",
            "as": "query_info"
        }
    },
    {
        "$unwind": "$insurance_info"
    },
    {
        "$unwind": "$query_info"
    }
]

# Thực hiện aggregation
results = collection_personal_info.aggregate(pipeline)

# Chuyển đổi ObjectId thành chuỗi và in kết quả
s_converted_results = [s_convert_objectid(result) for result in results]

# In kết quả
print("Dữ liệu kết nối từ các collection:")
print(json.dumps(s_converted_results, indent=4, ensure_ascii=False))
```
